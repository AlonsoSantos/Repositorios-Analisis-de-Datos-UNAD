# Kmeans

```python
# Realizar grÃ¡fica para identificar el valor de K

Nc = range(1, 8)
kmeans = [KMeans(n_clusters=i) for i in Nc]
scores = [kmeans[i].fit(Datos).inertia_ for i in range(len(kmeans))]  # Calcula la puntuaciÃ³n de cada modelo
plt.plot(Nc, scores, marker='o')
plt.xlabel('NÃºmero de clusters')
plt.ylabel('Score')
plt.title('GrÃ¡fica de codo')
plt.show()
```

```python
# generacion de modelo de agrupamiento de modelo con K=S (5 grupos de agrupamientos)
Modelo=KMeans(n_clusters=5,random_state=0)
Modelo.fit(Datos)
```

```python
# se agrega la columna Grupo a la base de datos para identificar el numero de grupo del agrupamiento

Modelo.labels_
Datos['Grupo']=Modelo.labels_
print(Datos)
```

```python
# desempeÃ±o del modelo con el coeficiente de Silhouette, indice de calinski-harabasz e indice de davies bouldin score

# preparacion de la metrica de la columna Grupo
Observaciones=len(Datos)
x=Datos.drop('Grupo',axis=1)
clusters=Datos['Grupo']

# calcular las metricas
sil_score=silhouette_score(x,clusters)
calinski_score=calinski_harabasz_score(x,clusters)
davies_score=davies_bouldin_score(x,clusters)

# crear tabla para mostrar valores de las metricas
table_data=[
    ['Numero de observaciones',Observaciones],
    ['Coeficiente Silhouette', sil_score],
    ['Indice Calinski-Harabasz',calinski_score],
    ['Indice Davies Bouldin',davies_score]
]

# mostrar tabla
from tabulate import tabulate
print(tabulate(table_data, headers=['Metric', 'Value'], tablefmt='pretty'))
```

# Hierarchical clustering

```python
dataset = pd.read_csv('./Mall_Customers.csv')
data = dataset.drop(['Gender'], axis=1)
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(data),columns=data.columns)
ward_clustering = linkage(X, method="ward", metric="euclidean")
plt.figure(figsize=(20,7))
dendrogram(ward_clustering)
plt.axhline(y=13, color='r', linestyle='-')
plt.title('Dendrograma')
plt.xlabel('Sample index')
plt.ylabel('Euclidean distance')
plt.show()
```

```python
hier_clustering = AgglomerativeClustering(n_clusters=3)
labels_hier = hier_clustering.fit_predict(X)
labels_hier
```

```python
# Obtener las etiquetas de cada punto de los datos
labels = labels_hier
colores=['red','green','cyan']
asignar=[]
for row in labels:
    asignar.append(colores[row])
```

```python
f1 = X['Age'].values
f2 = X['Spending Score (1-100)'].values
plt.scatter(f1, f2, c=asignar, s=70)
plt.title('Datos etiquetados')
plt.xlabel('Edad')
plt.ylabel('PuntuaciÃ³n de Gasto')
plt.show()
```

```python
def plot_dendrogram(model, **kwargs):
    '''
    Esta funciÃ³n extrae la informaciÃ³n de un modelo de agrupaciÃ³n aglomerativa
    y traza su dendrograma con la funciÃ³n dendrograma de scipy.cluster.hierarchy
    '''
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1
            else:
                current_count += counts[child_idx-n_samples]
            counts[i] = current_count
        distances = np.arange(model.children_.shape[0])
        linkage_matrix = np.column_stack([model.children_, distances, counts]).astype(float)
        plt.figure(figsize=(20,7))
        dendrogram(linkage_matrix, **kwargs)
        plt.title('Dendrograma')
        plt.xlabel('Sample index')
        plt.ylabel('Euclidean distance')
        plt.show()

plot_dendrogram(hier_clustering, color_threshold=0)
```

```python
# Metodo silhouette para identificar el numero optimo de clusters
range_n_clusters = range(2,15)
valores_medios_silhouette = []
for n_clusters in range_n_clusters:
    modelo = AgglomerativeClustering(affinity = 'euclidean',linkage = 'ward',n_clusters=n_clusters)
    clusters_labels = modelo.fit_predict(X)
    silhouette_avg = silhouette_score(X, clusters_labels)
    valores_medios_silhouette.append(silhouette_avg)
fig, ax = plt.subplots(1,1,figsize=(6,3.84))
ax.plot(range_n_clusters, valores_medios_silhouette, marker='o')
ax.set_title('Evolucion de media de los indices silhouette')
ax.xlabel('Numero clusters')
ax.ylabel('Media indices silhouette')
```
